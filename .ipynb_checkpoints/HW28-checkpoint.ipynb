```python
import numpy as np
import pandas as pd
```

#### Exercise 1

```python
df = pd.DataFrame({
'CustomerID': np.arange(1, 11),
'FirstName': ['John', 'Steve', 'Anna', 'Mike', 'Emily', 'Jake', 'Laura', 'Nick', 'Olivia',
'David'],
'LastName': ['Smith', 'Johnson', 'Williams', 'Jones', 'Brown', 'Davis', 'Miller',
'Wilson', 'Moore', 'Taylor'],
'Age': np.random.randint(25, 65, size=10),
'TotalPurchaseAmount': np.random.randint(100, 1000, size=10)
})
```

```python
df['classification'] = df.apply(lambda row: 'Low' if row['TotalPurchaseAmount'] < 500 else "High", axis = 1)
df.head()
```

```python
def classif(row) -> str:
    if row['classification'] == "Low" and row['Age'] < 20:
        return "Low_Young";
    elif row['classification'] == "High" and row['Age'] < 20:    
        return "High_Young";
    elif row['classification'] == "High" and row['Age'] >= 20:    
        return "High_Old";
    else:
        return "Low_Old";
    
df['classification_by_age'] = df.apply(lambda row: classif(row), axis=1)
df.head()
```

```python
sort_by_name = df.sort_values(by='FirstName')
sort_by_name.head()
```

```python
grouped = df.groupby('TotalPurchaseAmount').sum()
grouped.sort_values(['TotalPurchaseAmount', 'Age'], ascending=[True, False])
```
